```python
# One-Period Prisioners Dilemma

import numpy as np
import matplotlib.pyplot as plt
θ_path = []
θ = 0.4 # initial cooperation probability

def π(θ,a):
    return θ**a * (1-θ)**(1-a)

def dπ_dθ(θ,a):
    if a == 1:
        return 1
    else:
        return -1

def norm_r(r):
    return r - 2
        
rewards = np.array([[1, 5], 
                    [0, 3]])

α = 0.001
for i in range(1000):
    a = np.random.choice([0,1],p=np.array([1-θ,θ]))
    o_a = np.random.choice([0,1])
    r = rewards[a,o_a]
    θ = θ + α * norm_r(r) * dπ_dθ(θ,a) / π(θ,a)
    θ = np.clip(θ, 0, 1)
    θ_path.append(θ)

plt.plot(θ_path)
```

```python
# Infinite-Period Prisioners Dilemma with State
import numpy as np
import matplotlib.pyplot as plt
N = 1000
T = 20
G_path = []
θ0_path = []
θ1_path = []
θ = np.array([0.5, 0.5]) # cooperation probability if opponent defects and if opponent cooperates

def π(θ,a,s):
    if s == 0: # defection
        return θ[0]**a * (1-θ[0])**(1-a)
    elif s == 1: # cooperation
        return θ[1]**a * (1-θ[1])**(1-a)

def dπ_dθ(θ,a,s):
    if a == 1:
        return 1
    else:
        return -1

def norm_r(r,o_a):
    if o_a==0:
        return r - 2
    if o_a==1:
        return r - 2

rewards = np.array([[1, 5], 
                    [0, 3]])

α = 0.1
for i in range(N):
    s_init = np.random.choice([0,1])
    a_init = np.random.choice([0,1],p=[1-θ[s_init],θ[s_init]])

    for d in range(10):
        Gs = []
        o_s = np.random.choice([0,1])
        r_init = norm_r(rewards[a_init,o_s], o_s)
        s = o_s
        o_s = a_init
        G = r_init
        for t in range(T):
            a = np.random.choice([0,1],p=[1-θ[s],θ[s]])
            r = norm_r(rewards[a,o_s], o_s)    
            G += r
            s = o_s
            o_s = a
        Gs.append(G)
    G = np.mean(Gs)
            
    G_path.append(G)
    θ0_path.append(θ[0])
    θ1_path.append(θ[1])
    θ[s_init] = θ[s_init] + α * G/T * dπ_dθ(θ,a_init,s_init) / π(θ,a_init,s_init)
    θ = np.clip(θ, 0, 1)
    
plt.plot(G_path)
plt.show()
plt.plot(θ0_path, label='P(C) if opn D')
plt.plot(θ1_path, label='P(C) if opn C')
plt.legend()
plt.show()
```
