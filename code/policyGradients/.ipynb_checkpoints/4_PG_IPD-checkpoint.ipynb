```python
import sympy as sp

T = 5
R = 3
P = 1
S = 0
gamma = 0.95

# There are 4 possible states for memory-1 agents
# s in {dd,dc,cd,cc}, indexed by {1,2,3,4} 

# Policies: p(s),q(s) are probabilities of cooperating in different states
p1, p2, p3, p4, q1, q2, q3, q4, b = sp.symbols('p1 p2 p3 p4 q1 q2 q3 q4', real=True)

# Value functions corresponding to policies
v1, v2, v3, v4 = sp.symbols('v1 v2 v3 v4', real=True)

# Define bellman equations for policy
# V(s;p,q) = E[r(a,a')+Î²V(s';p,q)]
# s' = aa'
eq1 = v1 - (R*p1*q1 + T*(1-p1)*q1 + P*(1-p1)*(1-q1) + S*p1*(1-q1) + gamma*(v4*p1*q1 + v2*(1-p1)*q1 + v1*(1-p1)*(1-q1) + v3*p1*(1-q1)))
eq2 = v2 - (R*p2*q2 + T*(1-p2)*q2 + P*(1-p2)*(1-q2) + S*p2*(1-q2) + gamma*(v4*p2*q2 + v2*(1-p2)*q2 + v1*(1-p2)*(1-q2) + v3*p2*(1-q2)))
eq3 = v3 - (R*p3*q3 + T*(1-p3)*q3 + P*(1-p3)*(1-q3) + S*p2*(1-q3) + gamma*(v4*p3*q3 + v2*(1-p3)*q3 + v1*(1-p3)*(1-q3) + v3*p3*(1-q3)))
eq4 = v4 - (R*p4*q4 + T*(1-p4)*q4 + P*(1-p4)*(1-q4) + S*p2*(1-q4) + gamma*(v4*p4*q4 + v2*(1-p4)*q4 + v1*(1-p4)*(1-q4) + v3*p4*(1-q4)))

# Solve the equations
solutions = sp.solve([eq1, eq2, eq3, eq4], (v1, v2, v3, v4))

# Create derivative functions
derivative_p1 = sp.diff(solutions[v1], p1)
derivative_p2 = sp.diff(solutions[v2], p2)
derivative_p3 = sp.diff(solutions[v3], p3)
derivative_p4 = sp.diff(solutions[v4], p4)
derivative_q1 = sp.diff(solutions[v1], q1)
derivative_q2 = sp.diff(solutions[v2], q2)
derivative_q3 = sp.diff(solutions[v3], q3)
derivative_q4 = sp.diff(solutions[v4], q4)

# Create a dictionary of derivative functions
derivatives = {
    'dv1p1': sp.lambdify((p1, p2, p3, p4, q1, q2, q3, q4), derivative_p1, 'numpy'),
    'dv2p2': sp.lambdify((p1, p2, p3, p4, q1, q2, q3, q4), derivative_p2, 'numpy'),
    'dv3p3': sp.lambdify((p1, p2, p3, p4, q1, q2, q3, q4), derivative_p3, 'numpy'),
    'dv4p4': sp.lambdify((p1, p2, p3, p4, q1, q2, q3, q4), derivative_p4, 'numpy'),
    'dv1q1': sp.lambdify((p1, p2, p3, p4, q1, q2, q3, q4), derivative_q1, 'numpy'),
    'dv2q2': sp.lambdify((p1, p2, p3, p4, q1, q2, q3, q4), derivative_q2, 'numpy'),
    'dv3q3': sp.lambdify((p1, p2, p3, p4, q1, q2, q3, q4), derivative_q3, 'numpy'),
    'dv4q4': sp.lambdify((p1, p2, p3, p4, q1, q2, q3, q4), derivative_q4, 'numpy')
}

def computeDerivatives(s, p1, p2, p3, p4, q1, q2, q3, q4):
    args = [p1, p2, p3, p4, q1, q2, q3, q4]
    if s == 0:
        return derivatives['dv1p1'](*args),  derivatives['dv1q1'](*args)
    if s == 1:
        return derivatives['dv2p2'](*args),  derivatives['dv2q2'](*args)
    if s == 2:
        return derivatives['dv3p3'](*args),  derivatives['dv3q3'](*args)
    if s == 3:
        return derivatives['dv4p4'](*args),  derivatives['dv4q4'](*args)

def moving_average(x, w):
    return np.convolve(x, np.ones(w), 'valid') / w
    
```

```python
import numpy as np
import matplotlib.pyplot as plt
N = 5000 
alpha = 0.001
b = 0.99
cnt = 0
for i in range(100):
    p = np.random.uniform(0,1,(4,)) 
    q = np.random.uniform(0,1,(4,))
    p = np.clip(p/np.sum(p),0,1)
    q = np.clip(p/np.sum(q),0,1)
    p_init = p.copy()
    q_init = p.copy()
    
    payoff = np.array([[3, 0], [5, 1]])
    s = np.random.choice([0,1,2,3])
    rewards = []
    for period in range(N):
        # play prisioners' dilemma
        a1 = np.random.choice([0,1],p=[1-p[s],p[s]])
        a1_prime = np.random.choice([0,1],p=[1-q[s],q[s]])
        r1 = payoff[a1, a1_prime]
        r2 = payoff[a1_prime, a1]
        rewards.append(r1 + r2)
        
        # derivatives to Value function
        dVdp, dVdq = computeDerivatives(s, p[0], p[1], p[2], p[3], q[0], q[1], q[2], q[3])
        
        # Update p(s) and adjust other probabilities
        p[s] = p[s] + alpha * r1 * dVdp
        q[s] = p[s] + alpha * r2 * dVdq
        p = np.clip(p/np.sum(p),0,1)
        q = np.clip(p/np.sum(q),0,1)
        
        # Determine the state transition based on the actions played
        if a1 == 1 and a1_prime == 1:
            s = 3  # cc
        elif a1 == 1 and a1_prime == 0:
            s = 2  # cd
        elif a1 == 0 and a1_prime == 1:
            s = 1  # dc
        else:
            s = 0  # dd
    
    p_final = p.copy()
    q_final = q.copy()

    if np.mean(rewards[-100:])>5.75:
        cnt += 1
    
print(cnt)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def moving_average(x, w):
    return np.convolve(x, np.ones(w), 'valid') / w

N = 500 
alpha = 0.0018
b = 0.99
p_init = np.random.uniform(0, 1, (4,))
q_init = np.random.uniform(0, 1, (4))
payoff = np.array([[3, 0], [5, 1]])

def play_game():
    p = p_init.copy()
    q = q_init.copy()
    s = np.random.choice([0, 1, 2, 3])
    rewards = []

    for period in range(N):
        a1 = np.random.choice([0, 1], p=[1 - p[s], p[s]])
        a1_prime = np.random.choice([0, 1], p=[1 - q[s], q[s]])
        r1 = payoff[a1, a1_prime]
        r2 = payoff[a1_prime, a1]
        rewards.append(r1 + r2)

        dVdp, dVdq = computeDerivatives(s, p[0], p[1], p[2], p[3], q[0], q[1], q[2], q[3])

        p[s] = p[s] + alpha * r1 * dVdp
        q[s] = q[s] + alpha * r2 * dVdq
        p = np.clip(p / np.sum(p), 0, 1)
        q = np.clip(q / np.sum(q), 0, 1)

        if a1 == 1 and a1_prime == 1:
            s = 3  # cc
        elif a1 == 1 and a1_prime == 0:
            s = 2  # cd
        elif a1 == 0 and a1_prime == 1:
            s = 1  # dc
        else:
            s = 0  # dd

    return p, q, rewards

cooperation_data = []

for _ in range(100):  # Outer loop for sampling 100 times
    p_final, q_final, rewards = play_game()
    print(rewards[-10:])

    if np.mean(rewards[-100:]) > 5.1:
        cooperation_data.append((p_final, q_final, True))
    else:
        cooperation_data.append((p_final, q_final, False))

cooperation_p = []
cooperation_q = []
no_cooperation_p = []
no_cooperation_q = []

for p, q, reached_cooperation in cooperation_data:
    if reached_cooperation:
        cooperation_p.append(p)
        cooperation_q.append(q)
    else:
        no_cooperation_p.append(p)
        no_cooperation_q.append(q)

plt.scatter(cooperation_p, cooperation_q, c='b', label='Reached Cooperation')
plt.scatter(no_cooperation_p, no_cooperation_q, c='r', label="Didn't Reach Cooperation")

plt.xlabel('p')
plt.ylabel('q')
plt.legend()
plt.show()
```
