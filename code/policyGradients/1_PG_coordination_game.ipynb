### Initial Condition Effect

```python
import numpy as np
import matplotlib.pyplot as plt

for i in range(500):
    init_theta = np.random.uniform(0, 1, (2,)).reshape(2,1)
    K = 1
    alpha = 0.01 
    alpha_ = 0.01
    A = np.array([[1, (K + 1) * alpha], [(K + 1) * alpha_, 1]])
    B = np.array([-alpha, -alpha_]).reshape(2,1)
    theta = init_theta
    for j in range(1000):
        theta = np.clip(np.dot(A, theta) + B, 0, 1)
    
    if np.sum(np.round(theta)) == 0:
        plt.scatter(init_theta[0][0], init_theta[1][0], c='b', s=5)
    elif np.sum(np.round(theta)) == 2:
        plt.scatter(init_theta[0][0], init_theta[1][0], c='r', s=5)
    else:
        plt.scatter(init_theta[0][0], init_theta[1][0], c='y', s=5)

plt.xlabel('Theta 1')
plt.ylabel('Theta 2')
plt.title('2D Scatter Plot')
plt.show()
```
