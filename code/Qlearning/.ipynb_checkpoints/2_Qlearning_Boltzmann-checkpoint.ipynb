```python
def payoff(a1,a2):
    if a1>a2:
        return 1-a1
    elif a1==a2:
        return (1-a1)/2
    else:
        return 0

N = 2
action2bid = np.linspace(0,1,N+1)
R = np.zeros((N+1,N+1))
for i in range(N+1):
    for j in range(N+1):
        R[i,j] = payoff(action2bid[i],action2bid[j])

np.round(R,2)
R = R[0:N,0:N]
R
```

```python
import numpy as np

n_actions = N

alpha = 0.2
tau = 0.05
Q1 = np.ones(n_actions)*0.1
Q2 = np.ones(n_actions)*0.1

def select_action_boltzmann(Q, tau):
    p = np.clip(np.exp(Q / tau + 1e-8) / np.sum(np.exp(Q / tau + 1e-8)),1e-4,1-1e-4)
    p = p/np.sum(p)
    a = np.random.choice(n_actions, p=p)
    return a

a1_path = []
a2_path = []
r1_path = []
r2_path = []
n_episodes = 5000
for episode in range(n_episodes):
    a1 = select_action_boltzmann(Q1, tau)
    a2 = select_action_boltzmann(Q2, tau)
    r1 = R[a1, a2]
    r2 = R[a2, a1]
    Q1[a1] += alpha * (r1 + tau * np.max(Q1) - Q1[a1])
    Q2[a2] += alpha * (r2 + tau * np.max(Q2) - Q2[a2])
    a1_path.append(a1)
    a2_path.append(a2)
    r1_path.append(r1)
    r2_path.append(r2)
    if episode % (n_episodes/10) == 0:
        print(a1,a2,np.exp(Q1 / tau + 1e-8) / np.sum(np.exp(Q1 / tau + 1e-8)))
    tau = tau * 0.9999


def runningMeanFast(x, N):
    return np.convolve(x, np.ones((N,))/N)[(N-1):]

import matplotlib.pyplot as plt
plt.plot((runningMeanFast(a1_path,1000)))
plt.plot((runningMeanFast(a2_path,1000)))
plt.show()
plt.plot((runningMeanFast(r1_path,1000)))
plt.plot((runningMeanFast(r2_path,1000)))
plt.show()
```
